## 2. Đề xuất project (W26)

### Bài toán
Xây dựng một hệ thống trí tuệ nhân tạo (AI) có khả năng chơi cờ vua. Hệ thống có thể bao gồm:

- Chế độ chơi với AI: Người chơi đấu với AI có thể điều chỉnh độ khó.
- Chế độ AI đấu với AI: Quan sát cách hai AI thi đấu với nhau.
- Gợi ý nước đi: AI hỗ trợ người chơi bằng cách đề xuất nước đi tối ưu.
- Huấn luyện AI: AI có thể học từ các ván đấu trước đó để cải thiện chiến thuật.

### Phương pháp
1. Thuật toán tìm kiếm và đánh giá nước đi
- Minimax: Duyệt qua cây trạng thái của bàn cờ để tìm nước đi tối ưu.
- Alpha-Beta Pruning: Cải tiến Minimax để giảm số lượng trạng thái cần kiểm tra.
- Monte Carlo Tree Search (MCTS): Áp dụng cho AI học cách chơi bằng thử nghiệm nhiều ván đấu.

2. Mô hình học sâu (Deep Learning) để cải thiện AI
- Reinforcement Learning (Học tăng cường) với thuật toán như Deep Q-Learning hoặc AlphaZero-style training.

3. Xây dựng giao diện người dùng
- Sử dụng Python với thư viện Pygame để tạo giao diện.
